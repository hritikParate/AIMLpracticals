## BFS

```python
import collections
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}

def bfs(graph, root):

    visited, queue = set(), collections.deque([root])
    visited.add(root)

    while queue:

        # Dequeue a vertex from queue
        vertex = queue.popleft()
        print("\n"+str(vertex))


        # If not visited, mark it as visited, and
        # enqueue it
        for neighbour in graph[vertex]:
            if neighbour not in visited:
                visited.add(neighbour)

                print("neighbour is "+neighbour)
                queue.append(neighbour)

bfs(graph,'A')
```

## Water Jug Problem

```python
def pour_water(jug_A, jug_B):
  max1, max2, fill = 3,4,2
  print("%d\t\t\t%d" % (jug_A,jug_B))
  if jug_B == fill:
    return
  elif jug_B == max2:
    pour_water(0, jug_A)
  elif jug_A != 0 and jug_B == 0:
    pour_water(0, jug_A)
  elif jug_A == fill:
      pour_water(jug_A,0)
  elif jug_A < max1:
      pour_water(max1,jug_B)
  elif jug_A < (max2-jug_B):
      pour_water(0, (jug_A+jug_B))
  else:
    pour_water(jug_A-(max2-jug_B),(max2-jug_B)+jug_B)





print("Jug A 3ltrs \t Jug B 4ltrs")
pour_water(0,0)
```

### Feature Scaling using min max method

### new value = (old value - min)/(max-min)(new_max - new_min) + new_min

### new_max = 1, new_min=0
### new_value = (old_value - min of the column)/(column_max - column_min)(1-0)+0
